```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
import os

# sti = Path(os.getcwd())
# fil = sti.joinpath('05.json') 
# sti = Path(os.getcwd()).parent.parent
# fil = sti.joinpath('store_datafiler','05.json') 

with open("05.json") as f:
    df = pd.read_json(f)
topp_3 = df.value_counts('start_station_name').head(3)
bunn_3 = df.value_counts('start_station_name').tail(3)
s = pd.concat([topp_3, bunn_3])
sns.barplot(x=s.values, y=s.index, orient='h')
plt.show()
```
