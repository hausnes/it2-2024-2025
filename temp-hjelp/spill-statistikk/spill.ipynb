```python
# Fjern prosenttegn og konverter "Percentage_non_male" til numerisk verdi
df['Percentage_non_male'] = df['Percentage_non_male'].str.rstrip('%').astype('float')

# Sorter DataFrame etter "Percentage_non_male" i synkende rekkefølge
df_sorted = df.sort_values(by='Percentage_non_male', ascending=False)

# Hent de øverste radene
top_percentage_non_male = df_sorted[['Title', 'Percentage_non_male']].head(10).reset_index(drop=True)

# Vis resultatet
print(top_percentage_non_male)
```
